```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the dataset
data_frame = pd.read_csv("https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv")
```

```python
# Step 2: Display the first few rows of the dataset
data_frame.head()
```

```python
# Step 3: Generate summary statistics for the dataset
data_frame.describe()
```

```python
# Step 4: Data filtering and manipulation
# Select schools with names containing 'Newcomers' and 'Q' in their DBN
schools_with_newcomers_q = data_frame[data_frame['School Name'].str.contains('Newcomers') & (data_frame['School DBN'].str.contains('Q'))]
schools_with_newcomers_q
```

```python
# Select schools with names containing 'Newtown' and 'Q' in their DBN
schools_with_newtown_q = data_frame[data_frame['School Name'].str.contains('Newtown') & (data_frame['School DBN'].str.contains('Q'))]
schools_with_newtown_q 
```

```python
# Concatenate the two filtered datasets into one
merged_schools = pd.concat([schools_with_newcomers_q, schools_with_newtown_q], ignore_index=False)
merged_schools
```

```python
# Check for missing values in the merged dataset
missing_values_count = merged_schools.isnull().sum().sum()
missing_values_count
```

```python
# Get the column names of the merged dataset
merged_columns = merged_schools.columns
merged_columns
```

```python
# Step 5: Select desired columns for analysis
selected_columns = merged_schools[['School DBN', 'School Name', 'School Level', 'Regents Exam', 'Year','Total Tested', 'Mean Score', 'Number Scoring Below 65', 'Percent Scoring Below 65']]
selected_columns
```

```python
# Step 6: Further filtering
# Select rows where the 'Regents Exam' column contains 'Common Core Algebra2'
common_core_algebra2 = selected_columns[selected_columns['Regents Exam'].str.contains('Common Core Algebra2')]
common_core_algebra2
```

```python
# Step 7: Data transformation
# Convert the 'Number Scoring Below 65' column to a numeric data type, handling errors
numeric_scores_below_65 = pd.to_numeric(common_core_algebra2['Number Scoring Below 65'], errors='coerce')
numeric_scores_below_65
```

```python
# Step 8: Data grouping and aggregation
# Group data by the 'Year' column and calculate the count of students scoring below 65 for each year
yearly_scores_below_65 = numeric_scores_below_65.groupby(common_core_algebra2['Year']).count()
yearly_scores_below_65
```

```python
# Group data by school name and calculate the sum of students scoring below 65
sum_scores_below_65 = numeric_scores_below_65.groupby(common_core_algebra2['School Name']).sum()
sum_scores_below_65
```

```python
# Group data by school name and calculate the total number of students tested
total_students_tested = common_core_algebra2.groupby(common_core_algebra2['School Name'])['Total Tested'].sum()
total_students_tested
```

```python
# Step 9: Calculate the percentage of students scoring below 65
percentage_scores_below_65 = (sum_scores_below_65 / total_students_tested) * 100
percentage_scores_below_65
```

```python
# Step 10: Data visualization
# Create a bar plot to visualize the percentages of students scoring below 65 for each school
percentage_scores_below_65.plot(kind='bar', color='purple')
```

```python
percentages = [35, 16, 49]  # Replace with the actual percentages
school_names = ['Newtown High School', 'Newcomers High School', 'Other Schools']  # Replace with actual school names

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=school_names, autopct='%1.1f%%', colors=['purple', 'blue', 'green'])

# Adding a title
plt.title('Distribution of Students Scoring Below 65 by School')

# Display the pie chart
plt.show()
```

In summary, when we look at the students who took the Common Core Algebra2 test between 2016 and 2017, we see that 35% of Newtown High School students got scores below 65, while 16% of Newcomers High School students got scores below 65. This means that Newtown High School students struggled more compared to Newcomers High School students.
